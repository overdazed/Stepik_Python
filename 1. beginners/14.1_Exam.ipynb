**Star Triangle 🌶️**

Write a function ***draw_triangle()*** that outputs a star isosceles triangle with base and height equal **15 and 8** respectively:

```python
       *
      ***
     *****
    *******
   *********
  ***********
 *************
***************
```

**Note 1:** To draw a triangle, use a **for loop**. 

**Note 2:**  There are no spaces to the right of asterisks.

```python
# function declaration
def draw_triangle():
    # Define the base and height of the triangle
    base = 15
    height = 8
    
    # Loop through each row of the triangle
    for i in range(height):
        # Calculate the number of spaces needed before the stars in each row
        spaces = " " * (height - i - 1)
        # Calculate the number of stars needed in each row
        stars = "*" * (2 * i + 1)
        # Print the spaces followed by the stars for the current row
        print(spaces + stars)

# Call the function to draw the triangle
draw_triangle()
```

**Binomial coefficient 🌶️**

Write a function that takes two natural numbers and **compute_binom(n, k)** as arguments and returns the value of the binomial coefficient equal to *n k*

**(n!) / (k!(n-k)!)**

**Note 1.**  Factorial of a natural number *n*, is called the product of all natural numbers from 1 to *n*, that is **n!=1⋅2⋅3⋅…⋅n**

**Note 2.** Implement an auxiliary function **factorial(n)** that calculates the factorial of a number, or use a ready-made function from the module **math**. Please note that the function **compute_binom(n, k)** must return an integer. 

```python
def compute_binom(n, k):
    # Define a nested function to compute factorial
    def factorial(num):
        if num == 0:
            return 1
        else:
            return num * factorial(num-1)
    
    # Calculate the binomial coefficient using the factorial function
    result = factorial(n) / (factorial(k) * factorial(n - k))
    
    # Convert the result to an integer and return it
    return int(result)

# Take input for 'n' and 'k'
n = int(input())
k = int(input())

# Print the result of the binomial coefficient
print(compute_binom(n, k))
```

```python
from math import factorial as f

def compute_binom(n, k):
    return f(n) // (f(k) * f(n - k))


n = int(input())
k = int(input())

print(compute_binom(n, k))
```

```python
def number_to_words(num):
    # Define lists containing words for numbers before and after twenty
    before_twenty = [
        '', 'один', 'два', 'три', 'четыре', 'пять', 'шесть', 'семь', 'восемь', 'девять', 'десять', 'одиннадцать', 'двенадцать',
        'тринадцать', 'четырнадцать', 'пятнадцать', 'шестнадцать', 'семнадцать', 'восемнадцать', 'девятнадцать'
    ]
    after_twenty = [
        'двадцать', 'тридцать', 'сорок', 'пятьдесят',
        'шестьдесят', 'семьдесят', 'восемьдесят', 'девяносто'
    ]

    # If the number is less than 20, use the 'before_twenty' list directly
    if num < 20:
        res = before_twenty[num]
    else:
        # If the number is 20 or greater, use the 'after_twenty' list for tens place
        # and 'before_twenty' list for ones place
        res = after_twenty[num // 10 - 2] + " " + before_twenty[num % 10]

    return res.strip()

# Take input for the number
n = int(input())

# Print the number converted to words
print(number_to_words(n))
```

**Magic dates**

A magic date is the date when the day multiplied by the month equals the number formed by the last two digits of the year.

Write a function **is_magic(date)** that takes as an argument a string representation of a valid date and returns the value  ***True*** if the date is magic and ***False*** otherwise.

**Note.** The following program code:

- print(is_magic('10.06.1960'))
- print(is_magic('11.06.1960'))

should output:

- True
- False

```python
def is_magic(date):
    # Split the date string into day, month, and year
    parts = date.split('.')
    day, month, year = parts
    
    # Convert day, month, and year to integers
    d = int(day)
    m = int(month)
    y = int(year)
    
    # Extract the last two digits of the year
    last_two = int(str(y)[-2:])
    
    return d * m == last_two
    
date = input()

print(is_magic(date))
```

```python
def is_magic(date):
    seq = date.split(".")
    day, month, year = int(seq[0]), int(seq[1]), int(seq[2])
    
    return day * month == year % 100

date = input()

print(is_magic(date))
```

```python
def is_pangram(text):
    # Convert the text to lowercase to ensure case insensitivity
    text = text.lower()
    # Iterate through the ASCII values of lowercase letters from 'a' to 'z'
    for i in range(ord("a"), ord("z") + 1):
        if chr(i) not in text:
            # If any character is missing, return False (not a pangram)
            return False
    
    # If all characters are present, return True (pangram)
    return True

# Take input for the text
text = input()

# Print whether the text is a pangram or not
print(is_pangram(text))
```

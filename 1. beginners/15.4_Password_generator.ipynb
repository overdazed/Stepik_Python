```python
import random as rm

digits = '0123456789'
lowercase_letters = 'abcdefghijklmnopqrstuvwxyz'
uppercase_letters = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
symbols = '!#$%&*+-=?@^_'
ambi_chars = 'il1Lo0O'


chars = ''

_ask_number_passwords = 'How many passwords should I generate?'
_length = 'How long should the password/s be?'
_include_numbers = 'Should numbers be included? (y/n)'
_include_upper = 'Should capital letters be included? (y/n)'
_include_lower = 'Should lowercase letters be included? (y/n)'
_include_symbols = 'Should symbols be included? (y/n)'
_include_ambi = 'Should ambiguous characters "il1Lo0O" be included? (y/n)'

def password_data():
    cntPW = int(input(_ask_number_passwords))
    lenPW = int(input(_length))
    digON = input(_include_numbers)
    upperON = input(_include_upper)
    lowerON = input(_include_lower)
    symON = input(_include_symbols)
    ambiON = input(_include_ambi)
    
    if digON.lower() == 'y':
        chars += digits
    if upperON.lower() == 'y':
        chars += uppercase_letters
    if lowerON.lower() == 'y':
        chars += lowercase_letters
    if symON.lower() == 'y':
        chars += symbols
    if ambiON.lower() == 'n':
        for i in ambi_chars:
            chars = chars.replace(i, '')
            
    return cntPW, lenPW, chars

def generate_password(cntPW, lenPW, chars):
    for _ in range(cntPW):
        password = ''
        for _ in range(lenPW):
            password += rm.choice(chars)
        print(password)
        
def main():
    cntPW, lenPW, chars = password_data()
    generate_password(cntPW, lenPW, chars)
    
main()
```

### Password generation

Write a function ***generate_password()***, which takes two arguments:

- ***length***: password length
- ***chars***: the alphabet of characters the password consists of

and return the password.

Using a loop ***for***, generate the requeired number of passwords.

```python
from random import choice
from string import ascii_uppercase, ascii_lowercase, digits, punctuation

ambi_chars = 'il1Lo0O'
questions = ['Should numbers be included? (y/n)\n', 'Should uppercase letters be included? (y/n)\n', 'Should lowercase letters be included? (y/n)\n', 'Should symbols be included? (y/n)\n', 'Should ambiguous characters "il1Lo0O" be included? (y/n)\n']

answers = [digits, ascii_uppercase, ascii_lowercase, punctuation, ambi_chars]

count_password = int(input('How many passwords should be generated?\n'))
length_password = int(input('How long should the password be?\n'))

def request_param():
    chars = ''
    for i in range(len(questions) - 1): # Exclude the last question about ambiguous characters
        answer = input(questions[i]).lower()
        if answer == 'y':
            chars += answers[i]
    # Ask the last question separately
    ambi_answer = input(questions[4]).lower()
    if ambi_answer == 'n':
        for i in ambi_chars:
            chars = chars.replace(i, '')
            
    return chars

def generate_password(length_password, chars):
    lst_password = []
    for _ in range(count_password):
        password = ''
        while len(password) != length_password:
            password += choice(chars)
        lst_password.append(password)
    return lst_password

chars = request_param()
#print()
print(*generate_password(length_password, chars), sep='\n')
```

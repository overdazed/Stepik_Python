### __1. Check the nickname__

During the interview, you were asked to solve a username validation problem. The user is trying to create a nickname for his account on social network Y. The rules for a correct nickname on social network Y are as follows:

- nickname must begin with the @ symbol
- nickname must contain from 5 to 15(inclusive) characters
- nickname can only contain lowercase letters and numbers (besides the first @ symbol)

Write a program that prints "Correct" (without quotes) if the nickname matches all the rules above, otherwise the program should print "Incorrect" (without quotes).

```python
s = input()

if not s.startswith('@'):
    print('Incorrect')
elif not 5 <= len(s) <= 15:
    print('Incorrect')
elif not all(char.islower() or char.isdigit() for char in s[1:]):
    print('Incorrect')
else:
    print('Correct')
            
```

### __2. Bad comments 😈__

On the Stepik platform, users leave comments, but not all of them comply with the rules. For example, moderator Sam considers comments that are an empty line or consist only of spaces to be inappropriate. He deletes such comments - there is no point in littering the course with useless material!

Your task is to write a program that will help Sam check comments. The program must accept a natural number as input n, and then n lines representing comment texts. For each comment, your program should print the number of that comment (starting with 1), then a colon (:), then a space separated by its text or the message "COMMENT SHOULD BE DELETED" (without quotes) if the comment should be deleted by Sam.

**Input data format**
A natural number is given as input to the program n, and then n term.

**Output Format**
For each comment, the program should output its modified text or the message "COMMENT SHOULD BE DELETED" if the comment should be deleted.

```python
n = int(input())
cnt = 0
for i in range(n):
    s = input()
    cnt += 1
    if s.isspace() or s == '':
        print(str(cnt) + ': COMMENT SHOULD BE DELETED')
    else:
        print(str(cnt) + ': ' + s)
```

### __3. Car number 🚘__

The traffic service decided to optimize the process of creating license plates: instead of a person, they entrusted the generation of license plates to some GPT (machine learning model). As we know, artificial intelligence is still raw and makes many mistakes, so its results should be carefully checked. The correct license plate number (in Russia) has the following format:

https://ucarecdn.com/4b3309f6-b75f-469a-8fab-c7d3e3c83448/

Write a program that takes a string as input and checks whether the string is a valid license plate number. The program should output "YES" (without quotes) if the artificial intelligence completed its task, or "NO" (without quotes) otherwise. In our problem, we will consider the following formats to be the correct license plate number:

```python
<БУКВА><ЦИФРА><ЦИФРА><ЦИФРА><БУКВА><БУКВА>_<ЦИФРА><ЦИФРА>
<БУКВА><ЦИФРА><ЦИФРА><ЦИФРА><БУКВА><БУКВА>_<ЦИФРА><ЦИФРА><ЦИФРА>
```

```python
<LETTER><DIGIT><DIGIT><DIGIT><LETTER><LETTER>_<DIGIT><DIGIT>
<LETTER><DIGIT><DIGIT><DIGIT><LETTER><LETTER>_<DIGIT><DIGIT><DIGIT>
```

```python
plate = input()

cyrillic = 'АВЕКМНОРСТУХ'

if (
    plate[0] in cyrillic and
    plate[1:4].isdigit() and
    plate[4] in cyrillic and
    plate[5] in cyrillic and
    plate[6] == '_' and
    plate[7:].isdigit()
):
    print("YES")
else:
    print("NO")
```

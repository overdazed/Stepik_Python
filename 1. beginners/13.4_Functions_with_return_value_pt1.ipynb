## Return function

A return function is similar to a non-return function in that:

- it is a set of instructions that performs a specific task;
- When a function needs to be executed, it is called.

However, when a value-returning function completes, it returns a value to the part of the program that called it. The value returned from a function is used like any other: it can be assigned to a variable, displayed on the screen, used in a mathematical expression (if it is a number), etc.

**!!!**    A return function returns a value back to the part of the program that called it.

We have already encountered many functions that return values:

- function int() – converts a string to an integer and returns it;
- function float() – converts a string to a real number and returns it;
- function range() – returns a sequence of integers 0, 1, 2, ...;
- function abs() – returns the absolute value of a number (modulus of a number);
- function len() – returns the length of a string or list.

A function with a return value is written in exactly the same way as one without, but it must have an instruction return.

Here is the general format for defining a return function in Python:

```python
def function_name():
    code block
    # The function must contain an instruction returnthat takes the form:
    return expression
```

The value of the expression that follows the keyword *return* will be sent to the part of the program that called the function. This can be a variable or an expression, for example, a mathematical one.

**!!!** A return function has a statement returnthat returns a value to the part of the program that called it.

Let's write a function that performs the translation:

The purpose of this function is to take one number *temp* as an argument - the number of degrees on the Fahrenheit scale, and return another - the number of degrees on the Celsius scale.

Let's take a look at her work. The first statement in a function block assigns a value *(5 / 9) * (temp - 32)* to the variable *result*. ## Making the Most of the Return Statement

Let's take another look at the function *convert_to_celsius()*:

```python
def convert_to_celsius(temp):
    result = (5 / 9) * (temp - 32)
    return result
```

Notice that two things happen inside this function: first, the variable *result* is assigned the value of the expression *(5 / 9) * (temp - 32)*, and second, the value of the variable *result* is returned from the function. This function does its job well, but it could be simplified. Since the instruction *return* returns the value of the expression, *result* we eliminate the variable and rewrite the function like this:

```python
def convert_to_celsius(temp):
    return (5 / 9) * (temp - 32)
```

This version of the function does not store the value *(5 / 9) * (temp - 32)* in a separate variable, but returns the value of the expression immediately using the statement *return*. Does the same thing as the previous version, but in one step.

**Problem 1.**  Write a function that returns the length of the hypotenuse of a right triangle given the known values ​​of its legs.

**Solution 1**  To find the length of the hypotenuse, we need to apply the Pythagorean theorem: the square of the hypotenuse of a right triangle is equal to the sum of the squares of its legs. In other words, if a,b– lengths of the legs, and c is the length of the hypotenuse, then the equality holds:

c² = a²+b² ⟹ c= √(a²+b²)
 
A function that calculates the length of the hypotenuse can look like:

```python
def compute_hypotenuse(a, b):
    c = (a ** 2 + b ** 2) ** 0.5
    return c
```

**!!!** To extract the square root, we used the exponentiation operator. Recall that the result of both expressions: *math.sqrt(c)* and c**0.5 is one number √c

The following program code:

```python
print(compute_hypotenuse(3, 4))
print(compute_hypotenuse(5, 12))
print(compute_hypotenuse(1, 1))
```

If we need to pass numbers read from the keyboard to the program, then we write the following code:

```python
x = int(input())
y = int(input())

hypotenuse = compute_hypotenuse(x, y)

print(hypotenuse)
```

**Solution 3**  The function  sum_digits(n) might look like:

```python
def sum_digits(n):
    result = 0
    while n > 0:
        result += n % 10
        n //= 10
    return result
```

## Merging two sorted lists

Merging two sorted lists into one is an important problem in computer science. It naturally arises when sorting lists using **merge sort** https://en.wikipedia.org/wiki/Merge_sort. 

Let two lists of numbers *list1* and, sorted in ascending order, be given *list2*:

```python
list1 = [3, 10, 11, 12, 47, 57, 58, 63, 77, 79, 80, 95]
list2 = [0, 11, 12, 20, 24, 26, 47, 48, 53, 65, 70, 81, 84, 84, 90]
```

The simplest solution to the problem of merging lists uses the list method *sort()*:

```python
def merge(list1, list2):
    result = list1 + list2   # create the resulting list
    result.sort()            # sort the list using the built-in sort() method
    return result            # return the sorted list


list1 = [3, 10, 11, 12, 47, 57, 58, 63, 77, 79, 80, 95]
list2 = [0, 11, 12, 20, 24, 26, 47, 48, 53, 65, 70, 81, 84, 84, 90]
list3 = merge(list1, list2)  # call the function of merging two sorted lists

print(list3)
```

And although the function *merge()* completely copes with its task, it absolutely does not take into account the fact that the two lists *list1* and *list2* are already sorted.

## Quickly merge two sorted lists into one

Let us have two lists already **sorted in ascending order** *list1* and *list2*.

The fast merge algorithm is as follows:

1. We create numerical pointers *p1 = 0* to  *p2 = 0* the beginning of both lists *list1* and, *list2* respectively;
2. At each step we take the smaller of the two elements *list1[p1]* and *list2[p2]*;
3. We write it into the resulting list; 
4. Increase by 1 a pointer to the first element of the list (*p1* or *p2*) from which the element was taken;
5. When one of the initial lists is finished, we add all the remaining elements of the second list to the resulting list.

```python
def quick_merge(list1, list2):
    result = []

    p1 = 0  # pointer to the first element of list1
    p2 = 0  # pointer to the first element of list2

    while p1 < len(list1) and p2 < len(list2):  # before any of the lists run out
        if list1[p1] <= list2[p2]:
            result.append(list1[p1])
            p1 += 1
        else:
            result.append(list2[p2])
            p2 += 1

    if p1 < len(list1):   # residue clamping
        result += list1[p1:]
    else:                 # Otherwise, we attach the rest of the other list
        result += list2[p2:]
    
    return result
```

```python
list1 = [3, 10, 11, 12, 47, 57, 58, 63, 77, 79, 80, 95]
list2 = [0, 11, 12, 20, 24, 26, 47, 48, 53, 65, 70, 81, 84, 84, 90]
list3 = quick_merge(list1, list2)
print(list3)
```

#### __7. Merge lists 2__

The number is given as input to the program* n*, and then *n* lines containing integers **in ascending order**. Lists of numbers are formed from these lines. Write a program that combines the given lists into one sorted list using the function  quick_merge()and then prints it.

**Input data format**
A natural number is given as input to the program *n*, and then *n* strings containing integers in ascending order, separated by a space character.

**Output Format**
The program should output the elements of the final sorted list, each separated by a space.

```python
# Define a function called quick_merge that takes variable number of lists as input
def quick_merge(*lists):
    # Initialize an empty list to store merged elements
    merged_list = []
    # Iterate over each list in the input lists
    for lst in lists:
        # Extend the merged_list with the elements of the current list
        merged_list.extend(lst)
    # Sort the merged list and return
    return sorted(merged_list)

# Input: Read the number of lists (n) from the user
n = int(input())
# Initialize an empty list to store the input lists
lists = []

# Iterate n times to read n lists from the user
for i in range(n):
    # Read a line of input and split it into a list of strings
    line = input().split()
    # Convert each string element into an integer and append it to lists
    lists.append([int(num) for num in line])

# Combine and sort the lists using the quick_merge function
sorted_list = quick_merge(*lists)

# Output: Print the sorted list elements separated by space
print(*sorted_list)
```

```python
def quick_merge(list1, list2):
    result = []

    p1 = 0  # pointer to the first element of list1
    p2 = 0  # pointer to the first element of list2

    while p1 < len(list1) and p2 < len(list2):  # until at least one list runs out
        if list1[p1] <= list2[p2]:
            result.append(list1[p1])
            p1 += 1
        else:
            result.append(list2[p2])
            p2 += 1

    if p1 < len(list1):   # residue clamping
        result += list1[p1:]
    else:                 # Otherwise, we attach the rest of the other list
        result += list2[p2:]
    
    return result


n = int(input())
seq = []
for _ in range(n):
    seq.append([int(el) for el in input().split()])

while len(seq) >= 2:
    list1 = seq.pop(0)
    list2 = seq.pop(0)
    
    new_list = quick_merge(list1, list2)
    seq.append(new_list)

print(*seq[0])
```

### Functions that return multiple values

In Python, functions are not limited to returning just one value. After the statement, returnyou can define many expressions, separated by commas:

```python
return expression 1, expression 2, expression 3 ...
```

The following code defines a function get_powers(num) that takes a number as an argument num and returns its square, cube, and fourth power.

```python
def get_powers(num):
    return num**2, num**3, num**4
```

The result of executing the following code:

```python
a, b, c = get_powers(2)
print(a)
print(b)
print(c)
```

**!!!** Functions that return multiple values ​​are an exclusive feature of the Python language. Most programming languages ​​use a result data type, a list, that can contain multiple values ​​to return multiple values.

Let's look at another example. Suppose you need to write a function that finds the intersection point of two non-parallel lines *ax + by = e* and *cx + dy = f*. 
In other words, you need to solve the system of equations:

{*ax + by = e*
{*cx + dy = f*

It is not difficult to come to a solution to this system:



{*ax + by = e*     <=>   {x = (d*e - b*f) / (a*d - b*c)

{*cx + dy = f*     <=>   {y = (a*f - c*e) / (a*d - b*c)




The program code that solves the problem looks like:

```python
def solve(a, b, c, d, e, f):
    x = (d * e - b * f)/(a * d - b * c)
    y = (a * f - c * e)/(a * d - b * c)
    return x, y
```

The following code:

```python
xsol, ysol = solve(2, 3, 4, 1, 2, 5)
print('The solution to the system is the numbers', 'x =', xsol, 'y =', ysol)
```

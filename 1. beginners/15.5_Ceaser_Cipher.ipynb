### Cipher Ceaser

**The Caesar cipher** (shift cipher) is one of the simplest and most widely known encryption methods. A Caesar cipher is a type *of substitution cipher* (https://en.wikipedia.org/wiki/Substitution_cipher) in which each character in the plaintext is replaced by a character that is some constant number of positions to the left or right of it in the alphabet.

**!!!** Substitution cipher is an encryption method that replaces elements of the original plaintext with others, in accordance with a certain rule.

For example, in a right-shift cipher by 3 position, the symbol **A** is replaced by the symbol **D**, the symbol **B** is replaced by the symbol **B**, and so on, until the symbol **Z** is replaced by the symbol **C**.

https://commons.wikimedia.org/wiki/File:Caesar3.svg?uselang=ru

**!!!** The cipher is named after the Roman general *Gaius Julius Caesar*, who used it for secret correspondence with his generals.

The Caesar cipher is easy to crack and has almost no practical use.

### Mathematical model

If we compare each character of the alphabet with its serial number (numbering from 0), then encryption and decryption can be expressed by modular arithmetic formulas:


y = (x+k) mod n, 
x = (yв€’k) mod n,

Where *x*- plaintext symbol, and *y*- cipher text symbol, *n* is the power of the alphabet (https://de.wikipedia.org/wiki/M%C3%A4chtigkeit_(Mathematik)) (number of characters), and *k*- key.

**!!!** By operation  mod we mean the operation of finding the remainder. In Python, to find the remainder when dividing two numbers, we use the operator %.

### Code breaking

The Caesar cipher can be easily broken even if the attacker only knows the ciphertext. Two situations can be considered:

- 1. The cracker knows (or assumes) that a simple substitution cipher was used, but does not know that it is a Caesar scheme.
- 2. The cracker knows that a Caesar cipher was used, but does not know the shift value.
In the first case, the cipher can be cracked using the ***frequency analysis*** (https://en.wikipedia.org/wiki/Frequency_analysis) method . Using this method, an attacker is likely to quickly notice the regularity in the solution and realize that the cipher used is a Caesar cipher.

In the second case, breaking the cipher is even simpler. There are not many options for shift values вЂ‹вЂ‹(26 for English, 32 for the Russian language), all of them can be checked by brute force.

### Notes

**Note 1:** 
The transformation used in the Caesar cipher is usually denoted as ROT N, where N is the shift, ROT an abbreviation for the word ROTATE, in this case вЂњcyclic shiftвЂќ. 
For example, the notation ROT 2indicates a shift by 2 positions, that is, вЂњaвЂќ turns into вЂњcвЂќ, вЂњbвЂќ into вЂњdвЂќ, and so on, and at the end вЂњuвЂќ turns into вЂњaвЂќ and вЂњiвЂќ into вЂњbвЂќ.

**Note 2:** 
The number of different transformations depends on the length of the alphabet:
    - for Russian language it is possible 32 various transformations (transformations ROT 0 save ROT 33 the original text, and then repetitions begin);
    - possible for English 25 various transformations (transformations ROT 0 save ROT 26the original text, and then repetitions begin).

**Note 3:** 
Non-alphabetic characters - punctuation marks, spaces, numbers - do not change.

**Note 4:**  
A natural development of the Caesar cipher was the VigenГЁre cipher (https://en.wikipedia.org/wiki/Vigen%C3%A8re_cipher).

### Ave, Caeser

The input to the program is a string of text in English, in which all the words need to be encrypted. Each word of the string should be encrypted using a Caesar cipher (cyclic shift by the length of that word). Lowercase letters remain lowercase, and uppercase letters remain uppercase. It is guaranteed that there is one space between different words.

**Input data format**
- The program receives a line of text in English as input.

**Output format**
- The program must output ciphertext in accordance with the task conditions.

**Note.**  
Characters that are not English letters are not changed.

```python
# Define a function to encrypt a word using the Caesar cipher with a given shift
def caesar_encrypt(message, shift):
    # Initialize an empty string to store the encrypted word
    encrypted_word = ""
    # Iterate through each character in the message
    for char in message:
        # Check if the character is an alphabet letter
        if char.isalpha():
            # Convert character to ASCII Code
            # Determine the ASCII offset based on the case of the character
            ascii_offset = ord('a') if char.islower() else ord('A')
            # Apply the Caesar cipher encryption formula
            # like last letter Z = 97 = offset
            encrypted_char = chr((ord(char) - ascii_offset + shift) % 26 + ascii_offset)
            print(chr((ord(char))))
            # Append the encrypted character to the encrypted word
            encrypted_word += encrypted_char
        # If the character is not an alphabet letter
        else:
            # Append the character unchanged to the encrypted word
            encrypted_word += char
    # Return the encrypted word
    return encrypted_word

# Define a function to encrypt the entire text by encrypting each word using Caesar cipher with a shift equal to the word's length
def encrypt_text(text):
    # Split the input text into words using space as the delimiter
    # contains the individual words extracted from the input text
    words = text.split(' ')
    
    
    # Encrypt each word in the text using Caesar cipher with a shift equal to the word's length, and store the results in a list
    
    # 1. [...]: This denotes the start of a list comprehension, which is a concise way to create lists in Python.
    # list comprehension to iterate through each word in the list 'words'
    
    # 2. caesar_encrypt(word, sum(c.isalpha() for c in word)): 
    # This part is the expression that generates the elements of the list. 
    # For each word in the words list, it calls the caesar_encrypt function with two arguments: 
    # the word itself and the shift value calculated using sum(c.isalpha() for c in word). 
    # This calculates the shift as the total number of alphabetical characters in the word.
    # For each word, it calculates the shift for the Caesar cipher by counting the number of alphabetical characters in that word using sum(c.isalpha() for c in word).
    
    # 3. for word in words: 
    # This part specifies the iteration over each element in the words list. It iterates through each word in the input text.
    # Then, it calls the caesar_encrypt function to encrypt the word using the calculated shift. Finally, it stores the encrypted word in a list called encrypted_words.
    
    # 4. encrypted_words = [...]: 
    # This assigns the result of the list comprehension to the variable encrypted_words, effectively creating a list of the encrypted words.
    
    # So, overall, this line generates a list of encrypted words by iterating through each word in the input text, 
    # calculating the shift for each word based on the count of alphabetical characters, 
    # and encrypting the word using the Caesar cipher.
    encrypted_words = [caesar_encrypt(word, sum(c.isalpha() for c in word)) for word in words]
    # Join the encrypted words into a single string with spaces in between
    return ' '.join(encrypted_words)

text = input() # Read input text from the user
print(encrypt_text(text)) # Encrypt the input text and print the encrypted result 
```

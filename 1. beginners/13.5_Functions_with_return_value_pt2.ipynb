### Returning Boolean Values

Python allows you to write Boolean functions that return either true ( True), or false ( False). A Boolean function can be used to test a condition, then the values ​​True and False will signal that it is true.

Boolean functions are widely used to simplify complex conditions tested in decision structures and iterative structures.
For example, let's write a program that asks the user to enter a number and determines whether it is even or odd. 

```python
number = int(input())
if number % 2 == 0:
    print('This number is even.')
else:
    print('This number is odd.')
```

This piece of code will be easier to understand if you write it as a Boolean function *is_even()*, which takes a number as an argument and returns *True*, if it is even, and *False* if odd. 

```python
def is_even(number):
    if number % 2 == 0:
        return True
    else:
        return False
```

Now you can rewrite the instructions *if-else* main program so that it can determine the parity of a variable *number* called the function *is_even()*: 

```python
number = int(input())
if is_even(number):
    print('This number is even.')
else:
    print('This number is odd.')
```

This makes the program logic easier to understand, and the function can be called in the program whenever you need to check the parity of a number. 

### Using Boolean Functions to Validate Input Data

Boolean functions can also be used to **_simplify complex input validation code_** . For example, in a program that prompts the user to enter a product model number, where the only possible values ​​are 100, 200 or 300, we can write the following code: 

```python
model = int(input())

while model != 100 and model != 200 and model != 300:
    print('Valid model numbers are 100, 200, or 300.')
    model = int(input())
```

The validation loop uses a long compound Boolean expression and repeats until *model* will not equal 100, 200 or 300. 

At the same time, the validation cycle can be simplified by writing a Boolean function for checking the variable *model* and calling it in a loop. Let's write a function *is_invalid()*, which takes one parameter *model* and returns the value *True*, if the model is invalid and *False* otherwise. Then the validation loop can be rewritten as follows: 

```python
while is_invalid(model):
    print('Valid model numbers are 100, 200, or 300.')
    model = int(input())
```

After this change, the loop becomes easier to read. Now it is quite obvious that the cycle repeats until the model number is invalid. The code snippet below shows how the function could be written *is_invalid()*. It takes the model number as an argument, and if the argument is not equal 100, 200 or 300 then this function returns *True*, saying that it is unacceptable. Otherwise the function returns *False*. 

```python
def is_invalid(model):
    if model != 100 and model != 200 and model != 300:
        return True
    else:
        return False
```

**!!!** Creating functions that implement such simple logic is not always the optimal solution, since it increases the size of the code and leads to time spent calling the function and returning the result, which can negatively affect program performance.

**4. Good Password 🌶️**

Write a function *is_password_good(password)*, which takes the string password value as an argument passwordand returns the value *True* if the password is strong and *False*- otherwise. 

A password is strong if:

- its length is **not less 8** characters;
- it contains at least **one capital letter** (upper case);
- it contains at least **one lowercase letter** (lower case);
- it contains at least **one digit**.


```python
def is_password_good(password):
    # Check if the password length is less than 8 characters
    if len(password) < 8:
        return False
    
     # Initialize flags to track if uppercase letter, lowercase letter, and digit are found
    upper_found, lower_found, digit_found = False, False, False
    
    # Iterate through each character of the password
    for char in password:
        # Check if the character is uppercase
        if char.isupper():
            upper_found = True
        # Check if the character is lowercase
        elif char.islower():
            lower_found = True
        # Check if the character is a digit
        elif char.isdigit():
            digit_found = True
            
    # Return True only if all criteria are met
    return upper_found and lower_found and digit_found

txt = input()

print(is_password_good(txt))
```

**6. Palindrome 🌶️**

Write a function is_palindrome(text) that takes the string text as an argument and returns True if the specified text is a palindrome and False otherwise.

**Note 1.** A palindrome is a line that reads the same in both directions

**Note 2.** When checking, consider capital and small letters to be the same, and ignore spaces as well as the characters , . ! ? -.

```python
#REFERENCE

def is_palindrome(word):
    # Remove spaces and convert to lowercase
    word = word.replace(" ", "").lower()
    # Compare the word with its reverse
    return word == word[::-1]

word = input().strip()
if is_palindrome(word):
    print("YES")
else:
    print("NO")
```

```python
def is_palindrome(text):
    # Define characters to be removed
    chars_to_remove = (" ", ",", ".", "!", "?", "-")
    
    # Remove specified characters and convert to lowercase
    cleaned_text = text.lower() # Convert the text to lowercase
    for char in chars_to_remove:
        cleaned_text = cleaned_text.replace(char,"") # Remove each specified character
    
    # Check if the cleaned text is equal to its reverse
    return cleaned_text == cleaned_text[::-1] # Check if the cleaned text is a palindrome

# Test the function
txt = input()
print(is_palindrome(txt))
```

```python
def is_palindrome(num):
    # Check if the number is equal to its reverse string representation
    return str(num) == str(num)[::-1]

def is_prime(num):
    # Check if the number is a prime number
    if num < 2: # == 1 # If the number is less than 2, it's not prime
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0: # If the number is divisible by any number other than 1 and itself, it's not prime
            return False
    return True # If no such divisor is found, the number is prime

def is_even(num):
    # Check if the number is even
    return num % 2 == 0

def is_valid_password(password):
    # Split the password into its components
    parts = password.split(':')
    if len(parts) != 3: # If there are not exactly three parts, the password is invalid
        return False
    a, b, c = parts
    try:
        # Convert parts to integers
        a = int(a)
        b = int(b)
        c = int(c)
    except ValueError:
        # If conversion to integers fails, the password is invalid
        return False
    
    # Check if a is a palindrome, b is prime, and c is even
    if not is_palindrome(a): # If a is not a palindrome, the password is invalid
        return False
    if not is_prime(b): # If b is not prime, the password is invalid
        return False
    if not is_even(c): # If c is not even, the password is invalid
        return False
    
    return True # If all checks pass, the password is valid

# Get password input from the user
psw = input()
# Check if the password is valid and print the result
print(is_valid_password(psw))
```

Write a function *is_correct_bracket(text)* that takes as an argument a non-empty string text consisting of the characters "(" and ")" and returns *True* if the input string is a valid bracketed sequence and *False* otherwise.

**Note 1.** A valid parenthesis sequence is a string consisting only of the characters ( and ), where each opening parenthesis is paired with a closing parenthesis (each opening parenthesis must be to the left of its corresponding closing parenthesis).

**Note 2.** 
The following program code:
- print(is_correct_bracket('()(()())') 
- print(is_correct_bracket('()()(')))
should print:
- True 
- False

- 1. Each opening parenthesis ( must have a corresponding closing parenthesis ) appearing after it.
- 2. The number of opening and closing parentheses must be equal.
- 3. The opening and closing parentheses must occur in the correct order, meaning a closing parenthesis ) cannot appear before its corresponding opening parenthesis (.
- 4. An empty string is considered a valid parenthesis sequence.

```python
# function declaration
def is_correct_bracket(text):
    stack = []  # Initialize an empty list to act as a stack
    for char in text:
        if char == '(': # If the character is an opening parenthesis
            stack.append(char) # Push it onto the stack
        elif char == ')': # If the character is a closing parenthesis
            if not stack: # If the stack is empty, meaning no matching opening parenthesis
                return False # Return False as there's no matching opening parenthesis
            else:
                stack.pop() # Pop the last opening parenthesis from the stack, indicating a match
                # This line removes the last element (opening parenthesis) from the seq list (stack). 
                # Since we're using it in the context of checking parenthesis matching, 
                # it effectively simulates popping the last opening parenthesis from the stack 
                # when a closing parenthesis is encountered, indicating that the corresponding 
                # opening parenthesis has been matched.
    return len(stack) == 0 # If the stack is empty, all parentheses are matched, otherwise, return False

# read data
txt = input()

# call the function
print(is_correct_bracket(txt))
```

**9. Snake register**

Write a function that takes a **camel case** *convert_to_python_case(text)* string as an argument and converts it to **snake case**.

**Note 1.** You can read more about naming styles here .

**Note 2.** The following program code:

- print(convert_to_python_case('ThisIsCamelCased'))
- print(convert_to_python_case('IsPrimeNumber'))

should output:

- this_is_camel_cased
- is_prime_number

```python
def convert_to_python_case(text):
    snake_case = '' # Initialize an empty string for the snake_case version
    first_upper = True # Flag to track the first uppercase character
    
    # Iterate through each character in the input text
    for char in text:
        if char.isupper() and not first_upper: # If the character is uppercase and it's not the first one
            snake_case += '_' + char # Add underscore before the uppercase character
        else: 
            snake_case += char # Otherwise, add the character as it is
             # Check if the character is uppercase and if it's the first one encountered
            if char.isupper() and first_upper:
                first_upper = False # Update the flag to indicate that the first uppercase character has been encountered
            
    return snake_case.lower() # Convert the snake_case string to lowercase and return

txt = input()

print(convert_to_python_case(txt))
```

```python
def convert_to_python_case(text):
    new_text = ""
    for el in text:
        if not el == el.lower():  # check that the element is uppercase (skip digits)
            new_text += "_" + el.lower()
        else:
            new_text += el
            
    return new_text[1:]

txt = input()

print(convert_to_python_case(txt))
```

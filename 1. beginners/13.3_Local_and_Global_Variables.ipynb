## Local Variables

**Local** variables are those declared inside a function and accessible only to the function itself. Program code outside the function has no access to them.

Consider a function *print_texas()* that displays information about the number of birds living in Texas.

```python
def print_texas():
    birds = 5000
    print('Texas is home to', birds, 'birds.')
```

In the body of the function, we create a variable *birds* that is assigned a value equal to *5000*. Such a variable is local to the function *print_texas()*. Whenever a variable within a function is assigned a value, **a local variable is created** as a result . It belongs to the function in which it is created, and can only be accessed by that function's code.

**!!!** The term **“local”** indicates that the variable can only be used in this place - inside the function in which it is created.

If one function's code tries to access a local variable that belongs to another function, an error will occur.

```python
def print_texas():
    birds = 5000
    print('Texas is home to', birds, 'birds.')

def print_california():
    print('California is home to', birds, 'birds.')
```

A function *print_california()* accesses *birds* a function local variable *print_texas()*. Calling the function *print_california()* results in an error:

```python
NameError: name 'birds' is not defined
```

Local variables are hidden from other functions, so other functions can have their own local variables with the same name. For example,

```python
def print_texas():
    birds = 5000
    print('Texas is home to', birds, 'birds.')
print_texas()

def print_california():
    birds = 9000
    print('California is home to', birds, 'birds.')
print_california()
```

Each of these two functions has a local variable called *birds*. But they are never visible at the same time, since they are in different functions.

When a function is executed *print_texas()*, a variable is visible *birds* whose value is *5000*. When a function is executed *print_california()*, a variable is visible *birds* whose value is *9000*.

**!!!** Different functions can have local variables with the same names because they don't see each other's local variables.

## Variable Scope

The scope of a variable is the part of the program in which it can be accessed, the function where it was created. A variable is visible only to program code in its scope. No instruction outside the function can access such a variable.

A local variable cannot be accessed by program code that appears inside a function **before** the variable has been created.

For example, if *print_texas()* you swap lines of code in a function:

```python
def print_texas():
    print('Texas is home to', birds, 'birds.')
    birds = 5000
print_texas()
```

```python
# then when calling this function we will receive an error:
UnboundLocalError: local variable 'birds' referenced before assignment
```

The error resulted from a premature access to a local variable that had not yet been declared **birds**.

## Parametric Variable Scope

The scope of a parametric variable is the function in which the parameter is used. All program code for this function has access to the parametric variable.

Let's consider a function we already know:

```python
def draw_box(height, width):
    for i in range(height):
        print('*' * width)
    
draw_box(6, 9)
```

Parametric variables here *height, width*. One local variable is declared inside the function *i*.

## Notes

**Note 1**: A parametric variable is also local.

**Note 2**:  Memory for local variables is allocated for the duration of the execution of a given function in a special area called stack (https://en.wikipedia.org/wiki/Stack_(abstract_data_type)). When the function terminates, the memory is freed; the internal results of the function are not saved from one call to another.

## Global Variables

**Global** variables are those declared in the main program and available to both the program and all its functions.

Consider the following program code:

```python
birds = 5000    # глобальная переменная

def print_texas():
    print('Texas is home to', birds, 'birds.')

def print_california():
    print('California is home to', birds, 'birds.')
```

At the very beginning of the program, we create a global variable *birds* whose value is *5000*. Next we describe two functions that access a global variable. The result of executing the following code:

```python
print_texas()
print_california()
```

**!!!** Information exchange between the main program and functions is carried out only using function parameters and global variables.

A function can use any global variables except those with the same names as its local variables. If a local variable is declared in a function with the same name as one of the global ones, then this global variable becomes inaccessible in this function, and when specifying the variable identifier, the local variable of the function will be accessed, and not the global one of the same name.

Consider the following program code:

```python
birds = 5000    # глобальная переменная

def print_texas():
    birds = 1000
    print('Texas is home to', birds, 'birds.')

def print_california():
    birds = 7000
    print('California is home to', birds, 'birds.')
```

At the very beginning of the program, we create a global variable *birds* whose value is *5000*. Next we describe two functions that create local variables with the same name *birds*. Thus, when accessing a variable *birds* inside functions, the local variable will be accessed.

```python
print_texas()
print_california()
```

## Global variables are evil

Most programmers agree that we should limit our use of global variables or not use them at all. The reasons are as follows.

- **Global variables make it difficult to debug a program.** The value of a global variable can be changed by any instruction in the program file. If you find that a global variable is storing a bad value, you'll have to look for all the instructions that access it to determine where the bad value is coming from. This is not easy to do in a program with thousands of lines of code.
- **Functions that use global variables usually depend on these variables.** If you need to use such a function in another program, you will most likely have to redesign the function so that it does not rely on a global variable.
- **Global variables make the program difficult to understand.** A global variable can be modified by any instruction in the program. If you need to understand some part of the program that uses a global variable, you will have to learn about all the other parts of the program that access that global variable.

**!!!** In most cases, you should create variables locally and pass them as arguments to the functions that need to access them.

## Global Constants

Although the use of global variables should be avoided, it is acceptable to use global constants in a program.

A **global constant** is a global name that refers to an immutable value. Because the value of a global constant cannot be changed while the program is running, you do not have to worry about the potential dangers typically associated with the use of global variables.

Although Python does not allow you to create true global constants, you can simulate them using global variables. If a global variable is not declared using a keyword globalinside a function, then there is no way you can change the value of that global variable inside that function. 

**!!!** The mathematical module *math* defines two global variables, *math.pi* and *math.e*, to which the mathematical values ​​of the constants  *π = 3.14159265* and *e = 2.71828 are assigned*.

## Keyword global

If you want a statement inside a function to assign a value to a global variable, then an additional step is required. In this case, the global variable must be declared inside the function.

```python
def print_texas():
    # global variable
    global birds
    birds = 5000
    print('В Техасе обитает', birds, 'птиц.')

def print_california():
    print('В Калифорнии обитает', birds, 'птиц.')

print_texas()
print_california()
```

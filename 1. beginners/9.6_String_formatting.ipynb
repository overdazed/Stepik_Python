## format() method

Storing strings in variables is convenient, but it is often necessary to collect strings from other objects (strings, numbers, etc.) and perform the necessary manipulations with them. For this purpose, you can use the string formatting mechanism.

```python
birth_year = 1992
text = 'My name is Timur, I was born in ' + birth_year

print(text)
```

This code causes a runtime error because we are trying to add a number and a string:

```python
TypeError: can only concatenate str (not "int") to str
```

To solve such a problem, we can use the function str(), which converts a numeric value to a string.

Below code:

```python
birth_year = 1992
text = 'My name is Timur, I was born in ' + str(birth_year)

print(text)
```

This code works, but converting a number to a string every time is not very convenient. For more descriptive formatting, we can use the string method __format()__.

The previous code can be rewritten as:

```python
birth_year = 1992
text = 'My name is Timur, I was born in {}'.format(birth_year)

print(text)
```

We pass the necessary parameters to the method _format()_, and Python puts them in place of the curly braces {}- __placeholders__ . We can create as many placeholders as we want.

Below code:

```python
birth_year = 1992
name = 'Timur'
profession = 'math teacher'
text = 'My name is {}, I was born in {}, I work as a {}.'.format(name, birth_year, profession)

print(text)
```

For clarity and formatting flexibility, we can use a serial number in the placeholder:  _{0}_, _{1}_, _{2}_ etc. This number determines the position of the parameter passed to the method _format()_(numbering starts from zero):

```python
birth_year = 1992
name = 'Timur'
profession = 'math teacher'
text = 'My name is {0}, I was born in {1}, I work as a {2}.'.format(name, birth_year, profession)

print(text)
```

The parameter nameappears in the placeholder {0}, the parameter age- in the placeholder {1}, etc. We can use the same number in multiple placeholders or not at all, and we can also use numbers in different orders.

Below code:

```python
name = 'Timur'
city = 'Moscow'
text1 = 'My name is {0}-{0}-{0}!'.format(name, city)
text2 = '{1} is my city and {0} is my name!'.format(name, city)

print(text1)
print(text2)
```

## f-strings

The method _format()_ does a good job of formatting strings, but if there are a lot of parameters, the code may seem a little redundant.

Below code:

```python
first_name = 'Taylor'
second_name = 'Swift'
country = 'USA'
birth_date = '1989/12/13'
birth_place = 'West Reading, Pennsylvania'
text = '{0} {1} is a very famous singer from the {2}. She was born on {3} in {4}.'.format(first_name, second_name, country, birth_date, birth_place)

print(text)
```

Python 3.6 introduced a new type of string: f-strings. By prefixing the string with _f_, placeholders can include code, such as a variable name or any other expression. f-strings provide a clean and intuitive way to format strings.

The previous code can be written as:

```python
first_name = 'Taylor'
second_name = 'Swift'
country = 'USA'
birth_date = '1989/12/13'
birth_place = 'West Reading, Pennsylvania'
text = f'{first_name} {second_name} is a very famous singer from the {country}. She was born on {birth_date} in {birth_place}.'

print(text)
```

The placeholder _{first_name}_ is replaced by the value of the variable _first_name_, the placeholder _{last_name}_ is replaced by the value of the variable _last_name_, etc.

As already mentioned, in addition to variables in f-string placeholders, we can use expressions.

Below code:

```python
print(f'5 + 2 = {5 + 2}')
print(f'5 - 2 = {5 - 2}')
print(f'5 * 2 = {5 * 2}')
print(f'5 / 2 = {5 / 2}')
```

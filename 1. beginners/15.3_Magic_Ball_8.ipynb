### Main program loop

- 1. Organize a loop that will ask the user for data;
- 2. Ask the user a question;
- 3. Print a random answer using a function ***choice()*** passing a list ***answers*** as an argument;
- 4. Check with the user whether he wants to ask another question, if yes, then return to the main program loop, if not, print a message ***'Come back if you have any questions!'*** and end the program.

```python
# Welcome message
_welcome_message = 'Welcome, I am a magic 8 ball and I know the answer to any questions you may have!'
# Goodbye Message
_bye_message = 'Thank you for the interesting questions! See you.. :)'
# Wrong inputWould you like to have an answer for your deepest desired question?
_wrong_input_message = 'That is not a question!'

answers = ["Undoubtedly", "I think so", "Not clear yet, try again", "Don't even think about it",
"Foregone", "Most likely", "Ask later", "My answer is no",
"Without any doubts", "Good prospects", "Best not to tell", "Not to my knowledge",
"You can be sure of that", "Yes", "Concentrate and ask again", "Very doubtful"]

def welcome():
    print(_welcome_message)
    
def bye():
    print(_bye_message)
    
def ask_name(name):
    if name:
        hi = 'Hello, ' + name + '!'
        print(hi)
    else:
        nini = "You don't have a name?!\nAnyways - hello!"
        print(nini)

def is_valid_question(your_question):
    return your_question.strip().endswith('?')
        
def get_question(question):
    import random
    if is_valid_question(question):
        ques = ("Here is your answer:\n" + random.choice(answers))
        print(ques)
    else:
        print(_wrong_input_message)
        
def ask_again():
    user_input = input('Press "Y" to ask me something! Or "N" to stop.\n').lower()
    if user_input == 'y' or user_input == 'j':
        return True
    else:
        return False

def start():
    welcome()
    your_name = input('Type your name:\n').capitalize()
    ask_name(your_name)
    while True:
        your_question = input('What is your question?:\n')
        get_question(your_question)
        if not ask_again():
            break
    bye()
    
start()
```

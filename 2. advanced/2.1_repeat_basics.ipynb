**Body mass index**

Write a program to calculate and estimate a person's body mass index (BMI). BMI shows whether a person weighs more or less than normal for his height. A person's BMI is calculated using the formula:

IMT = weight(kg) / height(m) × height(m)    ,

where mass is measured in kilograms and height in meters.

A person’s weight is considered optimal if his BMI is between 18.5 and 25. If BMI is less 18.5, then the person is considered to be underweight. If the BMI value is greater 25, then the person is considered to be overweight.

The program should display "Optimal weight" if the BMI is between 18.5 and 25(inclusive). "Underweight" if BMI is less 18.5 and “Overweight” if the BMI value is greater 25.

**Input data format**
The program is given two numbers as input: the mass and height of a person, each on a separate line. All input numbers are real, use data type for them float.

**Output format**
The program must output text in accordance with the task conditions.

```python
weight = float(input())
height = float(input())

bmi = weight / (height * height)

# print(bmi)

if 18.5 <= bmi <= 25:
    print('Оптимальная масса')
elif bmi < 18.5:
    print('Недостаточная масса')
else: #bmi > 25:
    print('Избыточная масса')
```

**Number reversal**

Given a five-digit or six-digit natural number. Write a program that will reverse the order of its last five digits.

**Input data format**
One natural five-digit or six-digit number is given as input to the program.

**Output format**
The program must output the number that will be obtained as a result of the reversal specified in the problem statement. The number must be printed without any leading zeros.

```python
number = input()

# Check if the number is five or six digits long
if len(number) == 5:
    # Reverse the number as a string
    reversed_number = number[::-1]
    # Remove leading zeros and print
    print(int(reversed_number))
elif len(number) == 6:
    # Take the last five digits, reverse them as a string, and keep leading zeros
    reversed_five = number[1:][::-1]
    # Combine the first digit with the reversed last five digits
    result = number[0] + reversed_five
    # Remove leading zeros from the combined result and print
    print(int(result))
```

**Standard American Convention**

A natural number is given as input to the program. Write a program that inserts commas into a given number according to the standard American convention for commas in large numbers.

**Input data format**
A natural number is given as input to the program 𝑛 (0<n<10¹⁰⁰).

**Output format**
The program must output a number with commas in accordance with the task conditions.

```python
n = input()

n = n[::-1] # Reverse the string

# insert commas every three digits
n = ','.join(n[i:i+3] for i in range(0, len(n), 3))

n = n[::-1] # Reverse the string back to original order

print(n)
```

**Josephus' problem 🌶️🌶️**

npeople numbered from 1 to 𝑛, standing in a circle. They start to count, everyone 𝑘. The th person in a row is eliminated from the circle, after which the count continues with the person next to him. Write a program to determine the number of the person who will remain in the circle last.

**Input data format**
Two numbers are given as input to the program 𝑛 and 𝑘, written on separate lines.

**Output format**
The program should output one number - the number of the person who will remain last in the circle.

**Note 1.** You can read more about the classic problem of Josephus here .

**Note 2.** A visualization of the algorithm’s operation can be viewed here .



```python
n = int(input())

bin_n = ''

if n == 0:
    bin_n = '0'
else:
    # Convert the integer to binary
    while n > 0:
        bin_n = str(n % 2) + bin_n
        n = n // 2
        
# Print the binary representation

if len(bin_n) > 0:
    first_char = bin_n[0]
    rest_of_str = bin_n[1:]
    
    winning_bin_n = rest_of_str + first_char
else:
    winning_bin_n = bin_n
    
#print(winning_bin_n)


try:
    decimal_win_n = int(winning_bin_n, 2)
    print(decimal_win_n)
except ValueError:
    print("Invalid binary number")
```

```python
def josephus_problem(n, skip):
    # list comprehension
    # people = [i for i in range(1, n+1)]
    people = []
    for i in range(1, n+1):
        people.append(i)
    
    # the pop() function does not count the variable it just removed
    skip -= 1
    
    # use skip-variable (k) to create index
    # to loop back around, example n=11 and k=19
    # count to 11, loop back around and count another 8
    # % returns the remainder
    index = skip % len(people)
    
    while len(people) > 1:
        # runs continuesly, until our list of people is only one remaining
        # remove the perticular person at the position that we fount in index
        # pop() = removes the perticular index that we specified
        people.pop(index)
        # update it and remove the next index at skip-length
        # if sum ends up beeing greater than the length of people, we can loop back around to the beginning with %-Operator again
        index = (index + skip) % len(people)
    
    # return people of 0, or first position in our people list
    # because after we break out of a while loop there will only be one element in this people list
    return people[0]

n = int(input())
k = int(input())

print(josephus_problem(n, k))
```

```python
n = int(input())
k = int(input())


# holds the final result
res = 0

# iterate from 1 to n inclusive
for i in range(1, n+1):
    # adds the step count 'k' to the current result 'res'
    # '% i' Takes the modulus with 'i' to keep the result within the bounds of the current sequence size 'i'
    res = (res + k) % i

# The +1 adjustment is because the problem (likely related to the Josephus problem) is traditionally 1-based indexing, while the modulus operation naturally results in 0-based indexing.
print(res + 1)
```

```python
n = int(input())
k = int(input())

# create a list of people where each person is just a number
people = list(range(1, n + 1))
# variable for the current "counting" position
cur = 0

# While we have more than one person
while len(people) > 1:
    # start counting from the current cur position, increasing continuously
    for _ in range(k - 1):
        cur += 1
        # if the count goes past the last man,
        # then we put it on the first person
        if cur == len(people):
            cur = 0
            
    people.pop(cur)
    # if the last person was removed,
    # then we'll put the count on the first person
    if cur == len(people):
        cur = 0

# bring out the only person left
print(people[0])
```

**Back, forward and vice versa**

The input to the program is a text string of natural numbers. A list of numbers is formed from the elements of the string. Write a program that swaps adjacent elements of a list ( a[0] c a[1], a[2] c a[3]etc.). If there is an odd number of elements in the list, then the last one remains in its place.

Input format
The input to the program is a string of text containing natural numbers separated by spaces.

Output format
The program should output the modified list, separating its elements with one space. 

```python
numbers = [int(n) for n in input().split()]

# Loop through the list from the first element to the second last element
# Skip one element each time (i.e., go through all the even-indexed elements)
for i in range(0, len(numbers) - 1, 2):
    # Swap each element with its next one
    numbers[i], numbers[i+1] = numbers[i+1], numbers[i]
    
# Convert each number in the list to a string and join them with spaces
# Print the resulting string
print(' '.join([str(num) for num in numbers]))
```

**A shift in development**

The program is given a line of text from natural numbers. A list of numbers is formed from the elements of the string. Write a program of cyclic shift of the elements of the list to the right, when the last element becomes the first, and the rest are shifted by one position forward, in the direction of increasing the indices.

Format of input data
The program is entered by a line of text from the separated spaces of natural numbers.

Format of Output Data
The program should display elements of the modified list with a cyclic shift, separating its elements with a single space.

```python
# Take a line of input, split it into a list where each number is a separate element
numbers = input().split()

# Perform a cyclic shift to the right:
# The last element becomes the first, and all other elements move one position to the right
numbers = [numbers[-1]] + numbers[:-1]

# Print the list of numbers as a space-separated string
#print(' '.join( numbers))

print(*numbers)
```

```python
# Take a line of input, split it into a list where each number is a separate element
n = input().split()

cnt = 0

# For each number in the list
for i in range(len(n)):
    # Assume the number is unique
    is_unique = True
    
    # Check all previous numbers
    for j in range(i):
        if n[i] == n[j]:
            # It's not unique
            is_unique = False
            break
    # If it's unique, increment the count
    if is_unique:
        cnt += 1
        
print(cnt)
```

**Product of Numbers**

Write a program to determine whether a number is the product of two numbers from a given set. The program should output the result as “YES” or “NO”.

**Input Format** 
The first line contains a natural number n (0 < n < 1000) - the number of numbers in the set. In the following n lines, the integers that make up the set are entered (they can be repeated). Then follows an integer, which is or is not the product of two numbers from the set.

**Output Format** 
The program should output “YES” or “NO” in accordance with the task condition.

**Note 1.** 
A number from the set cannot multiply itself. In other words, the two factors must have different indices in the set.

**Note 2.** 
Use nested loops to solve the problem.


```python
# Read the number of elements in the set
n = int(input())

# Read the numbers in the set
numbers = [int(input()) for _ in range(n)]

# Read the number to be checked
product = int(input())

# Initialize the answer as "NO"
answer = "НЕТ"

# Loop through each pair of numbers in the set
for i in range(n):
    for j in range(i+1, n):
        # If the product of the pair equals the number to be checked
        if numbers[i] * numbers[j] == product:
            # Set the answer as "YES"
            answer = "ДА"
            # Break the inner loop
            break
        # If the answer is "YES", break the outer loop
    if answer == "ДА":
        break
   
# Print the answer     
print(answer)
```

```python
# Read the size of the sequence
size = int(input())

# Read the sequence of numbers
seq = [int(input()) for _ in range(size)]

# Read the number to be checked
number = int(input())

# Initialize a flag as False. This will be used to indicate if a pair is found
flag = False

# Loop through each pair of numbers in the sequence
for i in range(size):
    for j in range(size):
        # Check if the indices are different and if the product equals the number
        if i != j and seq[i] * seq[j] == number:
            # If such a pair is found, set the flag as True
            flag = True

# If the flag is True, print "ДА". Otherwise, print "НЕТ"
if flag:
    print("ДА")
else:
    print("НЕТ")
```
